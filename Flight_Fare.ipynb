```python
lis=data["Duration"]
```

```python
new_lis=[]
for i in lis:
    if len(i.split(" "))==1:
        if "m" in i:
            i="0h "+i
        elif "h" in i:
            i=i+" 00m"
    new_lis.append(i)      
```

```python
data["Duration"]=new_lis
```

```python
data["dur_hr"]=data["Duration"].str.split(" ").str[0].replace("h","",regex=True).astype(int)
data["dur_min"]=data["Duration"].str.split(" ").str[1].replace("m","",regex=True).astype(int)
```

```python
data
```

```python
data["Total_Stops"].unique()
```

```python
data["Total_Stops"]=data["Total_Stops"].replace(['non-stop', '1 stop', '2 stops', '3 stops', '4 stops'],[0,1,2,3,4])
```
